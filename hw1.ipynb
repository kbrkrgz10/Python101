##Ödev 1 
Kullanıcıdan 5 değer al.
Girilen değeri ve değerin tipini yazdır.

```
from ast import literal_eval
for x in range(0,5):
  value = input("Bana bir değer ver sana tipini döneyim:  ")
  print(f"Giriş yaptığın değer: {value} ")
  typeValue = "str"
  try:
    typeValue = type(literal_eval(value))
  except:
    pass

  print("Değerinin tipi-> {}".format(typeValue))
```
