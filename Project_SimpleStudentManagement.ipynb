```
class Student():
  def __init__(self):
    self.name = "kübra"
    self.surname = "akgözlüoğlu"
    self.courseList = []

  def CalculateGrade(midterm,final,project):
    grade = midterm * 0.3 + final * 0.5 + project * 0.2
    print("Grade: ",grade)
    result=""
    if grade >= 90: result = "AA"
    elif 70 <= grade < 90: result = "BB"
    elif 50 <= grade < 70: result = "CC"
    elif 30 <= grade < 50: result = "DD"
    else:
      result = "FF"
    
    if result == "FF":
      print("You are failure")
    else:
      print("Your point is ", result)
```

```
student = Student()

controlCount = 0
while controlCount<3:
  name = input("Please enter your name: ")
  surname = input("Please enter your surname: ")
  if name.lower() == student.name and surname.lower() == student.surname:
    print(f"Welcome {student.name.upper()} {student.surname.upper()}")
    break;
  controlCount +=1

if controlCount == 3:
  print("Please try again later...")
else:
  courseList = {1:"Python",2:"C#",3:"Java",4:"SQL",5:"R"}
  print("Your courses: " +str(courseList))
  try:
    chosenCourse = int(input("Please enter course number for exam: "))
    while chosenCourse < 1 or chosenCourse > 5:
      chosenCourse = int(input("Invalid number! Please enter a course number from list for exam: "))

    midterm = int(input(f"Please enter {courseList[chosenCourse]} midterm result: "))
    final = int(input(f"Please enter {courseList[chosenCourse]} final result: "))
    project = int(input(f"Please enter {courseList[chosenCourse]} project result: "))
    CalculateGrade(midterm,final,project)
  except ValueError as Error:
    print("Please enter int value",Error)
```
